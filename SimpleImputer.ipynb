```python
import pandas as pd
import numpy as np
```

```python
miles = pd.DataFrame({'farthest_run_mi': [50,62, np.nan, 100, 26,13 ,31, 56]})
```

```python
miles
```

```python
miles.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(strategy = 'mean')
```

```python
imp_mean.fit_transform(miles)
```

```python
imp_median = SimpleImputer(strategy = 'median')
```

```python
imp_median.fit_transform(miles)
```

```python
imp_mode = SimpleImputer(strategy = 'most_frequent') #for mode
```

```python
imp_mode.fit_transform(miles)
```

```python
imp_const = SimpleImputer(strategy = 'constant', fill_value = 13)
```

```python
imp_const.fit_transform(miles)
```

```python
names = pd.DataFrame({
    'names': ['ronaldo', 'messi', 'trent', np.nan, 'pedri', 'maradona']
})
```

```python
names
```

```python
imp_constant_cat = SimpleImputer(strategy = 'constant', fill_value = 'kroos')
```

```python
imp_constant_cat.fit_transform(names)
```

To show which data is specifically filled i.e previously null add_indicator is used.

```python
imp_mean_marked = SimpleImputer(strategy ='mean', add_indicator = True)
```

```python
imp_mean_marked.fit_transform(miles)
```

```python
df = pd.DataFrame(
    { 'Name' : ['Pessi', 'ronaldo', 'Neymar', 'Mbappe', np.nan, 'Rudiger', 'Rodri'],
      'Goals_scored_24': [6, 10, np.nan, 12, 9, 3, 13 ]  
})
```

```python
df
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
    (imp_constant_cat, ['Name']),
    (imp_mean, ['Goals_scored_24']),
     remainder = 'drop'


)
```

```python
ct.set_output(transform = 'pandas')
```

```python
df_pandas = ct.fit_transform(df)
```
